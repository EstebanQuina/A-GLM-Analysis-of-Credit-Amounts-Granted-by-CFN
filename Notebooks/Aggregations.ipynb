```python
# By looking at the dataset, we can drop the columns SUBSECTOR and ACTIVIDAD, since they are grouped by SECTOR. 
df = df.drop(columns=['SUBSECTOR', 'ACTIVIDAD'])
df.head()
```

```python
# Columns SUBSISTEMA and ENTIDAD have only one unique value, so we can drop them as well.
df = df.drop(columns=['SUBSISTEMA', 'ENTIDAD'])
df.head()
```

```python
# We can also drop the CANTON column, since it can be grouped by PROVINCIA.
df = df.drop(columns=['CANTON'])
df.head()
```

```python
# We can group provinces into regions, since there are many provinces and it would be difficult to analyze them individually.
# Coastal region: Guayas, Manabí, El Oro, Santa Elena, Los Ríos, Esmeraldas
# Sierra region: Pichincha, Azuay, Cañar, Loja, Tungurahua, Chimborazo, Cotopaxi, Imbabura, Carchi, Bolívar, Napo, Pastaza, Sucumbíos, Orellana
# Amazon region: Morona Santiago, Zamora Chinchipe
# Insular region: Galápagos
coast = ['GUAYAS', 'MANABÍ', 'EL ORO', 'SANTA ELENA', 'LOS RÍOS', 'ESMERALDAS', 'SANTO DOMINGO DE LOS TSÁCHILAS']
sierra = ['PICHINCHA', 'AZUAY', 'CAÑAR', 'LOJA', 'TUNGURAHUA', 'CHIMBORAZO', 'COTOPAXI', 'IMBABURA', 'CARCHI', 'BOLÍVAR']
amazon = ['MORONA SANTIAGO', 'ZAMORA CHINCHIPE', 'SUCUMBÍOS', 'ORELLANA', 'NAPO', 'PASTAZA']
insular = ['GALÁPAGOS']
df['REGION'] = df['PROVINCIA'].apply(lambda x: 'COSTA' if x in coast else ('SIERRA' if x in sierra else ('AMAZONÍA' if x in amazon else 'INSULAR')))
df.head()
```

```python
#We can substitute the Provincia column with the Region column, since it is more general and we can analyze the data by region instead of by province.
df = df.drop(columns=['PROVINCIA'])
df.head()
```

```python
#We can drop FECHA since it is not relevant for our analysis and it would be difficult to analyze the data by date.
df = df.drop(columns=['FECHA'])
df.head()
```

```python
# We group SECTOR into 4 categories (Primary: extraction, Secondary: manufacturing, Tertiary: services, Quaternary: knowledge/IT)
mapeo_sectores = {
    'AGRICULTURA GANADERÍA  SILVICULTURA Y PESCA': 'Primary',
    'AGRICULTURA, GANADERÍA,  SILVICULTURA Y PESCA': 'Primary',
    'EXPLOTACIÓN DE MINAS Y CANTERAS': 'Primary',
    'INDUSTRIAS MANUFACTURERAS': 'Secondary',
    'CONSTRUCCIÓN': 'Secondary',
    'SUMINISTRO DE ELECTRICIDAD GAS VAPOR Y AIRE ACONDICIONADO': 'Secondary',
    'SUMINISTRO DE ELECTRICIDAD, GAS, VAPOR Y AIRE ACONDICIONADO': 'Secondary',
    'DISTRIBUCIÓN DE AGUA; ALCANTARILLADO, GESTIÓN DE DESECHOS Y ACTIVIDADES DE SANEAMIENTO': 'Secondary',
    'COMERCIO AL POR MAYOR Y AL POR MENOR REPARACIÓN DE VEHÍCULOS AUTOMOTORES Y MOTOCICLETAS': 'Tertiary',
    'COMERCIO AL POR MAYOR Y AL POR MENOR; REPARACIÓN DE VEHÍCULOS AUTOMOTORES Y MOTOCICLETAS': 'Tertiary',
    'TRANSPORTE Y ALMACENAMIENTO': 'Tertiary',
    'ACTIVIDADES FINANCIERAS Y DE SEGUROS': 'Tertiary',
    'ACTIVIDADES INMOBILIARIAS': 'Tertiary',
    'ACTIVIDADES DE ALOJAMIENTO Y DE SERVICIO DE COMIDAS': 'Tertiary',
    'ACTIVIDADES DE SERVICIOS ADMINISTRATIVOS Y DE APOYO': 'Tertiary',
    'ACTIVIDADES DE ATENCIÓN DE LA SALUD HUMANA Y DE ASISTENCIA SOCIAL': 'Tertiary',
    'ENSEÑANZA': 'Tertiary',
    'OTRAS ACTIVIDADES DE SERVICIOS': 'Tertiary',
    'INFORMACIÓN Y COMUNICACIÓN': 'Quaternary',
    'ACTIVIDADES PROFESIONALES CIENTÍFICAS Y TÉCNICAS': 'Quaternary',
    'ACTIVIDADES PROFESIONALES, CIENTÍFICAS Y TÉCNICAS': 'Quaternary'
}
df['SECTOR'] = df['SECTOR'].map(mapeo_sectores)
df.head()
```

```python
# We translate to english the elements of the TIPO DE CREDITO column for better understanding. 
# PRODUCTIVO CORPORATIVO', 'PRODUCTIVO EMPRESARIAL', 'PRODUCTIVO PYMES', 'MICROCREDITO DE ACUMULACION AMPLIADA'
mapeo_creditos = {
    'PRODUCTIVO CORPORATIVO': 'Corporate',
    'PRODUCTIVO EMPRESARIAL': 'Business',
    'PRODUCTIVO PYMES': 'SMEs',
    'MICROCREDITO DE ACUMULACION AMPLIADA': 'Microcredit'
}
df['TIPO DE CREDITO'] = df['TIPO DE CREDITO'].map(mapeo_creditos)
df.head()
```

```python
 # We translate to english the elements of the TIPO DE OPERACION column for better understanding.
# 'FACTORING', 'CREDITO', 'CONTINGENTE'
mapeo_operaciones = {
    'FACTORING': 'Factoring',
    'CREDITO': 'Credit',
    'CONTINGENTE': 'Contingent'
}
df['TIPO DE OPERACION'] = df['TIPO DE OPERACION'].map(mapeo_operaciones)
df.head()
```

```python
# We translate ESTADO DE LA OPERACION column to english as well. 'ORIGINAL', 'REFINANCIADA', 'REESTRUCTURADA', 'NOVADA'
mapeo_estados = {
    'ORIGINAL': 'Original',
    'REFINANCIADA': 'Refinanced',
    'REESTRUCTURADA': 'Restructured',
    'NOVADA': 'Novated'
}
df['ESTADO DE LA OPERACION'] = df['ESTADO DE LA OPERACION'].map(mapeo_estados)
df.head()
```

```python
# We can join rows if they have the same values in all columns except for NÚMERO DE OPERACIONES, and MONTO OTORGADO, in which case we can sum these two columns.
df = df.groupby(['REGION', 'SECTOR', 'TIPO DE CREDITO', 'TIPO DE OPERACION', 'ESTADO DE LA OPERACION']).agg({'NÚMERO DE OPERACIONES': 'sum', 'MONTO OTORGADO': 'sum'}).reset_index()
df.head()
```

```python
df.info()
```

```python
# we can transform MONTO OTORGADO  with logarithm to reduce the skewness of the data and make it more normal, which can help with the analysis and modeling.
import numpy as np
df['MONTO OTORGADO'] = np.log(df['MONTO OTORGADO'])
df.head()
```

```python
# We change the name of the columns to english for better understanding and analysis.
df = df.rename(columns={'REGION': 'Region', 'SECTOR': 'Sector', 'TIPO DE CREDITO': 'Credit Type', 'TIPO DE OPERACION': 'Operation Type', 'ESTADO DE LA OPERACION': 'Operation Status', 'NÚMERO DE OPERACIONES': 'Number of Operations', 'MONTO OTORGADO': 'Log-Amount Granted'})
df.head()
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_cfn_volumencreditocontingente_2025_enero-septiembre.csv', index=False)
```
